## T-SNE plot for word vectors

```python
def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

## LDA: Topic Modelling

```python

# Helper function
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
        
# Tweak the two parameters below
number_topics = 20
number_words = 100
# Create and fit the LDA model
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
# Print the topics found by the LDA model
print("Topics found via LDA:")
print_topics(lda, count_vectorizer, number_words)
```

## Train Test Split

```python

def get_train_test_data(split=0.1):
    train_df['target']=train_df.target.apply(lambda x: x.strip())
    X_train, X_test,y_train,y_test = train_test_split(train_df['Business Description'],train_df['target'],test_size=split,stratify=train_df['target'],random_state=42)
    return X_train, X_test,y_train,y_test
```

## MultinomailNB with Tf-idf

```python
X_train, X_test,y_train,y_test = get_train_test_data()
nb = Pipeline([('vect', CountVectorizer(max_features=4000)),
               ('tfidf', TfidfTransformer()),
               ('clf', MultinomialNB())])
nb.fit(X_train, y_train)

y_pred = nb.predict(X_test)


MNB_accuracy = accuracy_score(y_pred, y_test)
MNB_precision = precision_score(y_pred,y_test,average='weighted')
MNB_recall = recall_score(y_pred,y_test,average='weighted')
MNB_f1 = f1_score(y_pred,y_test,average='weighted')
print('accuracy %s' % MNB_accuracy)
print('precision %s' % MNB_precision)
print('recall %s' % MNB_recall)
print('f1 score %s' % MNB_f1)
```

## GradientBoosting with Tf-idf

```python
X_train, X_test,y_train,y_test = get_train_test_data()
text_clf = Pipeline([('vect', CountVectorizer(max_features=4000)),
                     ('tfidf', TfidfTransformer(smooth_idf=True,use_idf=True,sublinear_tf=True)),
                     ('clf', GradientBoostingClassifier(n_estimators=50)),
                     ])
text_clf.fit(X_train, y_train)
predicted = text_clf.predict(X_test)
GBC_accuracy = accuracy_score(predicted, y_test)
GBC_precision = precision_score(predicted,y_test,average='weighted')
GBC_recall = recall_score(predicted,y_test,average='weighted')
GBC_f1 = f1_score(predicted,y_test,average='weighted')
print('accuracy %s' % GBC_accuracy)
print('precision %s' % GBC_precision)
print('recall %s' % GBC_recall)
print('f1 score %s' % GBC_f1)
```

## ANN with tf-idf

```python

def TFIDF(X_train, X_test,MAX_NB_WORDS=4000):
    vectorizer_x = TfidfVectorizer(max_features=MAX_NB_WORDS)
    X_train = vectorizer_x.fit_transform(X_train).toarray()
    X_test = vectorizer_x.transform(X_test).toarray()
    print("tf-idf with",str(np.array(X_train).shape[1]),"features")
    return (X_train,X_test)

def Build_Model_DNN_Text(shape, nClasses, dropout=0.3):
    """
    buildModel_DNN_Tex(shape, nClasses,dropout)
    Build Deep neural networks Model for text classification
    Shape is input feature space
    nClasses is number of classes
    """
    model = Sequential()
    node = 512 # number of nodes
    nLayers = 4 # number of  hidden layer
    model.add(Dense(node,input_dim=shape,activation='relu'))
    model.add(Dropout(dropout))
    for i in range(0,nLayers):
        model.add(Dense(node,input_dim=node,activation='relu'))
        model.add(Dropout(dropout))
    model.add(Dense(nClasses, activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    return model


X_train, X_test,y_train,y_test = get_train_test_data()
encoder = LabelEncoder()
encoder.fit(train_df['target'])
y_train = encoder.transform(y_train)
y_test = encoder.transform(y_test)

X_train_tfidf,X_test_tfidf = TFIDF(X_train,X_test)
model_DNN = Build_Model_DNN_Text(X_train_tfidf.shape[1], 11)
model_DNN.fit(X_train_tfidf, y_train,
                              validation_data=(X_test_tfidf, y_test),
                              epochs=50,
                              batch_size=32,
                              verbose=2)
predicted = model_DNN.predict(X_test_tfidf)
predicted = np.argmax(predicted,axis=1)

score = model_DNN.evaluate(X_test_tfidf, y_test,
                       batch_size=32, verbose=1)

DNN_accuracy = accuracy_score(predicted, y_test)
DNN_precision = precision_score(predicted,y_test,average='weighted')
DNN_recall = recall_score(predicted,y_test,average='weighted')
DNN_f1 = f1_score(predicted,y_test,average='weighted')
print('accuracy %s' % DNN_accuracy)
print('precision %s' % DNN_precision)
print('recall %s' % DNN_recall)
print('f1 score %s' % DNN_f1)
```

## Glove word embedding with Bi-lstm

```python
embedding_path = "/Users/subir/Downloads/glove/glove.6B.300d.txt"
# create the word2vec dict from the dictionary
def get_word2vec(file_path):
    file = open(embedding_path, "r")
    if (file):
        word2vec = dict()
        split = file.read().splitlines()
        for line in split:
            key = line.split(' ',1)[0] # the first word is the key
            value = np.array([float(val) for val in line.split(' ')[1:]])
            word2vec[key] = value
        return (word2vec)
    else:
        print("invalid file path")
w2v = get_word2vec(embedding_path)
```

```python
def get_tokens(sentence):
    tokens = sentence.split(' ')
    tokens = [token for token in tokens if (token not in stopwords)]
    return (tokens)

token_list = (train_df['Business Description'].apply(get_tokens))
```

```python
num_words = 2*len(token_list)
```

```python
X_train, X_test,y_train,y_test = get_train_test_data()
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(train_df.target.tolist())
y_train = encoder.transform(y_train)
y_train = utils.to_categorical(y_train, 11)
# y_train = y_train.reshape(-1,1)

y_test = encoder.transform(y_test)
y_test = utils.to_categorical(y_test, 11)
# y_test = y_test.reshape(-1,1)
print("y_test",y_test.shape)
print("y_train",y_train.shape)
```

```python
tokenizer = Tokenizer(num_words=4000)
tokenizer.fit_on_texts(token_list)
word_index = tokenizer.word_index
word_seq_train = tokenizer.texts_to_sequences(X_train)
word_seq_test = tokenizer.texts_to_sequences(X_test)
word_index = tokenizer.word_index
print("dictionary size: ", len(word_index))

#pad sequences
word_seq_train = pad_sequences(word_seq_train, maxlen=max_seq_len)
word_seq_test = pad_sequences(word_seq_test, maxlen=max_seq_len)
```

```python
# saving tokenizer for prediction purpose
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# get the embedding matrix from the embedding layer
vocab_size = len(tokenizer.word_index) + 1
embedding_matrix = zeros((vocab_size, 300))
for word, i in tokenizer.word_index.items():
    embedding_vector = w2v.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
len(np.unique(y_test)), y_train.shape[0]
```

```python
# model architecture

input = Input(shape=(max_seq_len,))
model = Embedding(vocab_size,300,weights=[embedding_matrix],input_length=max_seq_len)(input)
model =  Bidirectional (LSTM (512,return_sequences=True,dropout=0.50),merge_mode='concat')(model)
model = TimeDistributed(Dense(128,activation='relu'))(model)
model = Flatten()(model)
model = Dense(64,activation='relu')(model)
model = Dropout(0.5)(model)
output = Dense(11,activation='softmax')(model)
model = Model(input,output)
model.compile(loss='categorical_crossentropy',optimizer=Adam(), metrics=['accuracy'])
```

```python

class TextPreprocessing:

    def __init__(self, input_text):
        self.input_text = input_text

    def pos_tagging(self, text):
        doc = nlp(text)
        l = ['DATE', 'GPE']
        for ent in doc.ents:
            if ent.label_ in l:
                text = text.replace(ent.text, ent.label_)
        return text

    def tokenizer_and_pad(self):
        from keras.preprocessing.sequence import pad_sequences
        # loading
        with open('/Users/subir/Codes/Miscellaneous/AnalytixLab/Mixed - Digital - Text Classification-Python/tokenizer.pickle',
                'rb') as handle:
            tokenizer = pickle.load(handle)
        pos_tagged_text = self.pos_tagging(self.input_text)
        stemmed_text = " ".join([stemmer.stem(i) for i in pos_tagged_text.split()])
        lmtz_text = " ".join([lmtzr.lemmatize(i) for i in stemmed_text.split()])
        ip_word_seq = tokenizer.texts_to_sequences([lmtz_text])
        ip_word_seq = pad_sequences(ip_word_seq, maxlen=150)
        return ip_word_seq
    
    
encoding_df = pd.read_csv(
    '/Users/subir/Codes/Miscellaneous/AnalytixLab/Mixed - Digital - Text Classification-Python/encodings.csv')


class Model:
    def __init__(self,model):
        self.model=model

    def pred_using_bilstm(self, ip_seq):
#         model = load_model(
#             "/Users/subir/Codes/Miscellaneous/AnalytixLab/Mixed - Digital - Text Classification-Python/ModelCheckpoint/BiLSTM_Fasttext.23-0.711.hdf5")

        encoding = np.argmax(model.predict(ip_seq.reshape(1, -1)))

        pred_label = encoding_df.loc[encoding_df.encoding == encoding, 'label'].values[0]

        return pred_label
```
